## PyCitySchools
* This report will help the school board and mayor make strategic decisions regarding future school budgets and priorities. It consists of analysis of the district-wide standardized test results of every student's math and reading scores, as well as various information on the schools they attend. It aggregates the data and showcase obvious trends in school performance.

```python
# Dependencies and Setup
import os
import pandas as pd
import numpy as np

school_data_to_load = os.path.join("Resources", "schools_complete.csv")
student_data_to_load = os.path.join("Resources", "students_complete.csv")

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Overview table that summarizes key metrics about each District, including:
    * Total number of schools
    * Total number of students
    * Total budget
    * Average math score 
    * Average reading score
    * Overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
    * Percentage of students with a passing math score (70 or greater)
    * Percentage of students with a passing reading score (70 or greater)



```python
new_school_df = school_data_complete.groupby(['school_name'])
```

```python
total_no_of_schools = new_school_df['school_name'].nunique().sum()
```

```python
total_number_of_students = new_school_df['student_name'].count().sum()
```

```python
total_budget = new_school_df['budget'].unique().sum()
# converting the total_budget into string type and adding a $ symbol before it
total_budget = '$'+ total_budget[0].astype(str)
```

```python
average_math_score = new_school_df['math_score', 'reading_score'].mean().mean()
```

```python
overall_passing_score = average_math_score.mean()
```

```python
percent_math_greater_70 = school_data_complete[school_data_complete['math_score']>=70][['math_score']].count()*100/school_data_complete['math_score'].count()
# Rounding the data to two decimal places, converting the data into string type and adding a % symbol after it
percent_math_greater_70 = percent_math_greater_70[0].round(2).astype(str)+'%'
```

```python
percent_reading_greater_70 = school_data_complete[school_data_complete['reading_score']>=70][['reading_score']].count()*100/school_data_complete['reading_score'].count()
# Rounding the data to two decimal places, converting the data into string type and adding a % symbol after it
percent_reading_greater_70 = percent_reading_greater_70[0].round(2).astype(str)+'%'
```

```python
District_summary = pd.DataFrame({\
                                 'Total schools': total_no_of_schools,\
                                 'Total students': total_number_of_students,\
                                 'Total budget' : [total_budget],\
                                 'Average Math Score': average_math_score['math_score'].round(2),\
                                 'Average Reading Score' : average_math_score['reading_score'].round(2),\
                                 'Overall passing rate' : overall_passing_score.round(2),\
                                 '% passing math' : [percent_math_greater_70],\
                                 '% passing reading' : [percent_reading_greater_70]\
                                })   
District_summary
```

## School Summary

* Overview table that summarizes key metrics about each school, including:
    * School Name
    * School Type
    * Total Students
    * Total School Budget
    * Per Student Budget
    * Average Math Score
    * Average Reading Score
    * % Passing Math
    * % Passing Reading
    * Overall Passing Rate (Average of the above two)

```python
df_1 = new_school_df['school_name','type','size', 'budget'].apply(lambda x: x).drop_duplicates()
school_data_df = pd.DataFrame({\
                     'school_name' : df_1['school_name'],\
                     'School Type' : df_1['type'],\
                     'Total Students' : df_1['size'],\
                      'Total Budget'  : ('$' + df_1['budget'].astype(str))\
                    })
school_data_df.set_index(['school_name'], inplace = True)
```

```python
school_score_avrg_df = new_school_df.agg({'math_score': 'mean',\
                   'reading_score' : 'mean',\
                    'budget': 'mean',\
                    'size': 'mean'\
                    }) 

school_score_avrg_df["Per student budget"] = ('$' + (school_score_avrg_df['budget']/school_score_avrg_df['size']).astype(str))

school_score_avrg_df = pd.DataFrame({'Budget per student': school_score_avrg_df['Per student budget'],\
                                    'Average Math score': school_score_avrg_df['math_score'],\
                                    'Average Reading score':school_score_avrg_df['reading_score']\
})
```

```python
percent_passing_math_school = (school_data_complete[school_data_complete['math_score']>=70][['math_score', 'school_name']].groupby(['school_name']).count()['math_score']/df_1.set_index('school_name')['size'])*100
percent_passing_math_school = pd.DataFrame(percent_passing_math_school, columns = ['% Passing Math'])
```

```python
percent_passing_reading_school = (school_data_complete[school_data_complete['reading_score']>=70][['reading_score', 'school_name']].groupby(['school_name']).count()['reading_score']/df_1.set_index('school_name')['size'])*100
percent_passing_reading_school = pd.DataFrame(percent_passing_reading_school, columns = ['% Passing reading'])
```

```python
school_score_passing_df = percent_passing_math_school.merge(percent_passing_reading_school, on = 'school_name')
school_score_passing_df["% Over all Passing"] = school_score_passing_df.mean(axis = 1)
school_score_passing_df
```

```python
school_score_avrg_df = school_score_avrg_df.merge(school_score_passing_df, on = 'school_name')
```

```python
school_summary = school_data_df.merge(school_score_avrg_df, on = 'school_name')
school_summary.rename_axis('School Name')
school_summary
```

## Scores by School Spending

* A table that breaks down school performances based on average Spending Ranges (Per Student). 
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 605, 625, 645,665]
group_names = ["<$585", "$585-605", "$605-625", "$625-645", "$645-665"]
```

```python
# removing the $ symbol and converting the data into float because this same dataFrame is utilized further for calculation
school_summary['Budget per student'] = school_summary['Budget per student'].map(lambda x: x.lstrip('$-'))
school_summary['Budget per student'] = school_summary['Budget per student'].astype(float)
```

```python
bins = pd.cut(school_summary['Budget per student'], spending_bins)
```

```python
performance_by_bin = school_summary.groupby(bins)['Average Math score','Average Reading score','% Passing Math', '% Passing reading','% Over all Passing'].mean()
```

```python
performance_by_bin.reset_index().set_index(pd.Index(group_names)).rename_axis('Spending Ranges(Per Student)').drop(['Budget per student'], axis =1)
```

## Scores by School Size

* Same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
size_group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
bins_size = pd.cut(school_summary['Total Students'], size_bins)
performance_by_size_bin = school_summary.groupby(bins_size)['Average Math score','Average Reading score','% Passing Math', '% Passing reading','% Over all Passing'].mean()
performance_by_size_bin.reset_index().set_index(pd.Index(size_group_names)).rename_axis('Total Students').drop(['Total Students'], axis =1)
```
